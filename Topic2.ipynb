# $\color{violet}{\text{Module 2: Understanding Relationships Between Variables}}$

In this worksheet, we will start investigating relationships between variables. We will continue working with our TikTok dataset.

The dataset includes the following variables. Descriptions of the variables are provided by the poster on Kaggle.

- **track_name**: the title of the track
- **artist_name**: the artist name
- **artist_pop**: popularity of the artist scored between 0 and 100, with 0 being the lowest popularity and 100 being the highest popularity.
- **album**: The name of the album
- **track_pop**: popularity of the song scored between 0 and 100, with 0 being the lowest popularity and 100 being the highest popularity.
- **danceability**: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
- **energy**: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.
- **loudness**: The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Values typically range between -60 and 0 db.
- **mode**: Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
- **key**: The key the track is in. Integers map to pitches using standard Pitch Class notation. If no key was detected it was assigned a -1.
  - 0 = C
  - 1 = C♯/D♭
  - 2 = D
  - 3 = D♯/E♭
  - 4 = E
  - 5 = F
  - 6 = F♯/G♭
  - 7 = G
  - 8 = G♯/A♭
  - 9 = A
  - 10 = A♯/B♭
  - 11 = B
- **speechiness**: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.
- **acousticness**: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
- **instrumentalness**: Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.
- **liveness**: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
- **valence**: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
- **tempo**: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.
- **time_signature**: An estimated time signature. The time signature is a notational convention to specify how many beats are in each bar. The time signature ranges from 3 to 7 indicating time signatures of "3/4", to "7/4".
- **duration**: The duration of the track in milliseconds.

citation: Svetlana C. (2022). Tiktok popular songs 2022, Retrieved May 1 2025 from https://www.kaggle.com/datasets/sveta151/tiktok-popular-songs-2022/data.

# $\color{cyan}{\text{Box Plots}}$

Great work relabeling some of our categories! Next, we will use box plots to understand how these categorical variables relate to some of our other variables.

**What is a Box Plot?**

- A Box plot shows how spread out the data is.
- You can have more than one box if there are multiple categories.
- The box shows the middle of the data.
  - The bottom of the box is the lower quartile (where 25% of the data is below).The top of the box is the upper quartile (where 75% of the data is below) This means that 50% of the data is in the box.
  - The line inside is the median (the middle value).
  - The whiskers go to the farthest data points that are not outliers.
  - The individual points are the outliers. This just means these points are quite far from the mediam (middle).
- The bigger the box, the higher that spread, or standard deviation, of the data.
- If you have multiple boxes for different categories, you can assess which category has more spread out data (bigger boxes) and which category has larger or smaller values based on how high or low the box is.

## $\color{olive}{\text{Code for Box Plots}}$

- We will use the ggplot2 library in R to make our box plots. Let's load in the library.




```r
library(ggplot2)
```

The command we will be using to make a box plot is:

- ggplot(dataset, aes(x = column of categories, y = columnfory)) +
  geom_boxplot(fill = "skyblue", color = "darkblue") +
  labs(title = "Your title here") +
  theme_minimal()
- - We can change the colours of our histogram. The colours available in ggplot can by found by clicking [this link](https://sape.inf.usi.ch/quick-reference/ggplot2/colour)

To start, lets make a box plot of the energy variable. Here I am not plotting by category, so we will only use the y= in the aes() function. I will not incorporate x=.

```r
ggplot(TikTok, aes(y = energy)) + geom_boxplot(fill = "mediumorchid", color = "darkblue") + labs(title = "Energy Box Plot") + theme_minimal()
```

- From this plot we can see that the median is around energy is around 0.2.
- 75% of the data is below 0.8ish, and 25% of the data is below 0.5ish.
- There is one outlier with a value close to 0.2.


I would like to know what some of the lowest energy songs are. I can use the filter function in R to see which songs have an energy that is less than 0.3.
  - This function is in the tidyverse library, so I will load that in first.

```r
library(tidyverse)
filter(TikTok, TikTok$energy < 0.3)
```

I have found my outlier with the low energy! It is the song Just the Two of Us!

## $\color{olive}{\text{Box Plots by Category}}$

I would like to know if the energy of the songs tends to be different depending on the key of the song. I can use a box plot to help me answer this question.

```r
ggplot(TikTok, aes(x = key,y = energy)) + geom_boxplot(fill = "mediumorchid", color = "darkblue") + labs(title = "Energy Box Plot") + theme_minimal()
```

It looks like the spread of the songs' energy changes depengding on the key of the song.
- Songs that are in Key G seem to have a very small spread in the energy.
- Songs that are in key A seem to have a very large spread in the energy.
- The medians/middles seems to be pretty similar, but the highest medians are in the Fs/Gf key and B key. The lowest median is in the As/Bf group.
- The boxes all somewhat overlap, so it does not seem like there is a strong relationship between key and energy.

It looks like there is one song in the B key that has an energy quite a but lower than the rest of the songs in B key. I am going to use the filter function in R to find this. song.

```r
filter(TikTok, energy < 0.4 & key == "B")
```

Therefore I am is a song in key B that has a lower energy level than most other popular TikTok songs in key B.

# $\color{cyan}{\text{Relationships Between Continuous Variables}}$

- A continuous variable is something you can measure, and it can have any number — even numbers with decimals.
- In our dataset there are many continuous variables:
  - track_pop, danceability, energy, loudness, speechiness,acousticness, instrumentalness, tempo, duration
- We be interested in finding relationships between some of these continuous variables.
- We can investigate these relationships using scatter plots!
  - A scatter plot is like a picture made of dots that shows how two things are connected.


Let's look at the relationship between energy and loudness using a scatter plot. We will be using the following code to make a scatter plot using the ggplot2 library.

ggplot(data, aes(x = columnforx, y = columnfory)) +
  geom_point(color = "blue") +
  labs(title = "Title") +
  theme_minimal()

```r
ggplot(TikTok, aes(x = energy, y = loudness)) +
  geom_point(color = "blue", size = 3) +
  labs(title = "Loudness Vs. Energy") +
  theme_minimal()
```

From this plot we can see that as energy increases, it looks like loudness increases as well! We would say that these variables look to have a positive relationship.

Notice that the dots do not follow a perfect straight line - this is because there are many other factors that effect loudness and enegy - they are not perfectly related, but there is a relationship!
